# Series

```python
import pandas as pd
```

###  Series: It is used for Single dimension Data

```python
x=[3,4,5,6,7]
var=pd.Series(x)
print(var)
```

```python
#you can any index not only 0,1,2,...
# Could set data type to be float of x, insted of int64
# Name of this series frame
var=pd.Series(x,index=['a','s','d','f','g'],dtype='float',name='Series1')
print(var)
```

```python
# Dictionary can also be Passed
dic={'name':['python','c','c++','java'],'por':[12,13,14,15],'rank':[1,4,3,2]}
var1=pd.Series(dic)
var1
# dtype:object ,because it is mixed data
```

```python
# If you want value at multiple indicies
s=pd.Series(12,index=[1,2,3,4,5])
s
```

### When can you use Series instead of Pandas

```python
# Adding two series together will add compenent wise and remaining extra element of one series adding to nothing of another series will give NAN
s1=pd.Series(12,index=[1,2,3,4,5,6,7,8])
# s1=[12,12,12,12,12,12,12,12]
s2=pd.Series(12,index=[1,2,3,4,5])
# s2=[12,12,12,12,12]
s1+s2
```

# Merge

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2,3,4],'C':[21,22,23,24]})
pd.merge(var1,var2,on='A') # on option for common col
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2,3,5],'C':[21,22,23,24]}) # 1,2,3,5 changed here
pd.merge(var1,var2,on='A') # the part which are same,for them only the data will be merged (here 1,2,3)
```

```python
pd.merge(var1,var2,how='left',indicator=True)# to include that is different and not taken by default (here 4 in var1 and 5 in var2)
# how='inner' will do nothing, its a default option
# how='right' will take right varible var2
# how='outer' will give 4 and 5
# indicator=True will show which is common data,which is present in left or right variable
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
pd.merge(var2,var1,left_index=True,right_index=True,suffixes=['_name','_python']) # add suffix instead of x and y
```

# Join

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'C':[10,20,30,40],'D':[11,22,33,44]})
var1.join(var2)
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'C':[10,20],'D':[11,22]})
var1.join(var2)
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['1','2','3','4'])
var2=pd.DataFrame({'C':[10,20],'D':[11,22]},index=['1','2'])
var1.join(var2)
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['1','2','3','4'])
var2=pd.DataFrame({'C':[10,20],'D':[11,22]},index=['1','2'])
var2.join(var1) # var2 inplace of var1
# if var2 has less value, joining will give small dataframe
```

```python
var2.join(var1,how='left') # how='right' # will result same as if you interchange var1 and var2 in above example
```

```python
# how='outer' will show union of data(will show all the data)
# how= 'inner' will show intersection(common of two)
var2.join(var1,how='inner')
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['1','2','3','4'])
var2=pd.DataFrame({'C':[10,20],'B':[11,22]},index=['1','2']) # if same col name B
var2.join(var1,lsuffix='_1',rsuffix='_1') # executing without suffix will give value error because of same col name.lsuffix(left suffix using left variable (var2)
```

# Group by

```python
import pandas as pd
var=pd.DataFrame({'name':['a','b','c','a','b','a','b','a','c'],
                   'S_1':[12,13,14,12,15,16,12,18,19],
                   's_2':[26,34,24,22,25,36,22,38,29]})
var
```

```python
var_new=var.groupby('name')
for x,y in var_new:
  print(x)
  print(y)
  print()
```

```python
var_new.get_group("a") # for specific data
```

```python
var_new.min()
```

```python
var_new.mean()
```

```python
li=list(var_new) # convert group data to list
li
```

# Pivot

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],
                  "st_name":["a","b","c","a","b","c"],
                  "eng":[10,11,12,13,14,15],
                  "maths":[17,181,19,13,16,17]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

```python
import pandas as pd
var=pd.DataFrame({"days":[1,1,1,1,2,2],
                  "st_name":["a","b","c","a","b","c"],
                  "eng":[10,11,12,13,14,15],
                  "maths":[17,181,19,13,16,17]})
var
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean") # aggfunc="sum" 
```

```python
var.pivot_table(index="st_name",columns="days",aggfunc="mean",margins=True)
```
